```python
num = int(input("Enter number: "))

rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10

print("Reverse:", rev)
```

```python
num = int(input("Enter number: "))

original = num
sum = 0

while num > 0:
    digit = num % 10
    sum += digit ** 3
    num //= 10

if original == sum:
    print("Armstrong Number")
else:
    print("Not Armstrong")
```
